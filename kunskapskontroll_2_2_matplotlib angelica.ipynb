### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
plt.plot(x, y)
plt.title('y = x^2')  # text
plt.xlabel('x')  # x-axis

plt.ylabel('y')  ## y-axis
plt.show()  ## plot show
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
np.random.seed(0)
scatter_x = np.random.randn(100)
scatter_y = np.random.randn(100)

```

```python
categories = ['Category A', 'Category B', 'Category C']
values = [25, 40, 35]
# subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
# 1 subplot: Scatter plot
ax[0].scatter(scatter_x, scatter_y, color='blue', alpha=0.6)
ax[0].set_title('Scatter Plot')
ax[0].set_xlabel('X values')
ax[0].set_ylabel('Y values')

# 2 subplot: Bar plot
ax[1].bar(categories, values, color='green')
ax[1].set_title('Category Distribution')
ax[1].set_xlabel('Category')
ax[1].set_ylabel('Values')

# visa plot
plt.tight_layout()
plt.show()
```
