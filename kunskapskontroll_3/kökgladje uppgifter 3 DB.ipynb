```python
# totalt sale per produkt , lager * gånger pris per sales
df['TotalSales'] = df['Quantity'] * df['PriceAtPurchase']

# grupp ProductID och totalt sales från högsta till lägsta
total_sales_per_product = df.groupby('ProductID')['TotalSales'].sum().sort_values(ascending=False)

print(total_sales_per_product)
```

```python
# grupp StoreID och totalt sales från högsta till lägsta 
sales_per_store = df.groupby('StoreID')['TotalSales'].sum().sort_values(ascending=False)

print(sales_per_store)
```

```python
# Column TransactionDate som data type datetime om det är inte så tidigare 
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])


# grupp TransactionDate totalt sales per månad 
sales_over_time = df.groupby(df['TransactionDate'].dt.to_period('M'))['TotalSales'].sum()
monthly_sales = df.groupby(['Month', 'ProductID'])['TotalPrice'].sum().reset_index()


print(sales_over_time)
```
